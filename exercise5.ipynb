```python
def caesar_decrypt(text, shift):
    decrypted = ""
    for char in text:
        if char.isalpha():
            shifted = ord(char) - shift
            if char.islower():
                if shifted < ord('a'):
                    shifted += 26
            elif char.isupper():
                if shifted < ord('A'):
                    shifted += 26
            decrypted += chr(shifted)
        else:
            decrypted += char
    return decrypted

encrypted_text = "Kqz}jNylyvc@(jGZ)jG-(jme"
shift_value = 3  # Örneğin, 3 kaydırma
decrypted_text = caesar_decrypt(encrypted_text, shift_value)
print(decrypted_text)

```

```python
def caesar_cipher_decrypt(text, shift):
    decrypted_text = ""

    for char in text:
        if char.isalpha():  # Eğer harfse
            shift_base = 65 if char.isupper() else 97  # Büyük harf veya küçük harf
            new_char = chr((ord(char) - shift_base - shift) % 26 + shift_base)
            decrypted_text += new_char
        else:
            decrypted_text += char  # Harf olmayan karakterler olduğu gibi eklenir

    return decrypted_text

# Şifreli metin
cipher_text = "Kqz}jNylyvc@(jGZ)jG-(jme"
shift_value = 3  # Kaydırma değeri

# Şifreyi çözme
decrypted_message = caesar_cipher_decrypt(cipher_text, shift_value)
print("Çözülen Mesaj:", decrypted_message)
```

```python
import zipfile
import os

# Dosya yolu (Google Drive'dan veya Colab'dan doğrudan yüklediyseniz uygun yolu yazın)
zip_file_path = '/content/drive/MyDrive/matryoshka.zip'  # Google Drive'dan
# zip_file_path = '/content/dosya.zip'  # Eğer Colab'a doğrudan yüklendi ise

# Çıkartma işleminin yapılacağı klasör
extract_path = '/content/extracted_files'

# Hedef klasörü oluşturun
os.makedirs(extract_path, exist_ok=True)

# ZIP dosyasını açın
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # 5000 kez çıkarma işlemi
    for i in range(5000):
        # Her seferinde çıkarılacak klasörün adını oluşturun
        subfolder_path = os.path.join(extract_path, f"extracted_{i+1}")
        os.makedirs(subfolder_path, exist_ok=True)

        # Dosyaları çıkartın
        zip_ref.extractall(subfolder_path)

        print(f"{i+1}. kez çıkarıldı: {subfolder_path}")

print("5000 kez çıkarma işlemi tamamlandı.")
```
