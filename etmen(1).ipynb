```python
from mesa import Agent, Model
import random

class Etmen(Agent):
    def __init__(self, model, unique_id):
        super().__init__(model)
        self.unique_id = unique_id
        self.varlik = random.randint(1, 10)

    def step(self):
        # basit dinamik: varlık artsın
        self.varlik += random.choice([-1, 0, 1])
        self.varlik = max(self.varlik, 0)

class Cevre(Model):
    def __init__(self, N):
        super().__init__()
        self.etmenler = []

        for i in range(N):
            self.etmenler.append(Etmen(self, i))

    def step(self):
        random.shuffle(self.etmenler)
        for etmen in self.etmenler:
            etmen.step()
```

```python
model = Cevre(3)

print("MODEL BAŞLADI")
model.step()
print("MODEL BİTTİ")
```

```python
yeni_model = Cevre(10)

for i in range(10):
    yeni_model.step()
```
