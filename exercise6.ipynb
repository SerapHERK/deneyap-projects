```python
binary_string = "0101001101101001011000100110010101110010010101100110000101110100011000010110111001111011001100010101111101110110011001010101111100110000010111110110010001110101011011100111100101100001011100110110100101111101"

# İkili string'i byte dizisine dönüştürme
byte_array = [binary_string[i:i+8] for i in range(0, len(binary_string), 8)]
bytes_object = bytes([int(b, 2) for b in byte_array])

# Bytes nesnesini UTF-8 ile decode etme
text = bytes_object.decode("utf-8")

print(text)
```

```python
def sezar_sifrele(metin, kaydirma):
    sifreli_metin = ""
    for char in metin:
        if char.isalpha():  # Eğer karakter harfse
            # Harfin ASCII kodunu al, kaydırma ekle ve tekrar harfe dönüştür
            base = 65 if char.isupper() else 97
            sifreli_metin += chr((ord(char) - base + kaydirma) % 26 + base)
        else:
            # Harf olmayan karakterleri olduğu gibi bırak
            sifreli_metin += char
    return sifreli_metin

# Kullanıcıdan şifreleme yapılacak metin ve kaydırma miktarı alınır
metin = "Kqz}jNylyvc@(jGZ)jG-(jme"
kaydirma = 3 # Örneğin 3 harf kaydırma

sifreli_metin = sezar_sifrele(metin, kaydirma)
print(f"Şifreli metin: {sifreli_metin}")
```

```python
def sezar_sifre_coz(metin, kaydirma):
    cozulmus_metin = ""
    for char in metin:
        if char.isalpha():  # Eğer karakter harfse
            base = 65 if char.isupper() else 97
            cozulmus_metin += chr((ord(char) - base - kaydirma) % 26 + base)
        else:
            cozulmus_metin += char  # Harf olmayan karakterleri olduğu gibi bırak
    return cozulmus_metin

# Verilen metin ve kaydırma miktarı
metin = "HryviEzgzm{4gy4hs_D4h_S3i3}"
kaydirma = 3  # Sezar şifrelemesindeki kaydırma miktarı

cozulmus_metin = sezar_sifre_coz(metin, kaydirma)
print(f"Çözülmüş metin: {cozulmus_metin}")
```
