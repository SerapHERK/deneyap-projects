```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import confusion_matrix
from sklearn import metrics
import seaborn as sns
import matplotlib.pyplot as plt

veri = pd.read_csv("hayvanatbahcesi.csv", encoding='unicode_escape')

# İSİM SÜTUNUNU KALDIR
girisler = veri.drop(["sinifi", veri.columns[0]], axis=1)
cikis = veri["sinifi"]

X_train, X_test, y_train, y_test = train_test_split(
    girisler,
    cikis,
    test_size=0.20,
    random_state=110
)

# X encode
enc = OrdinalEncoder()
X_train_enc = enc.fit_transform(X_train)
X_test_enc = enc.transform(X_test)

# y encode
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

# Model
gnb = CategoricalNB()
gnb.fit(X_train_enc, y_train_enc)
y_pred = gnb.predict(X_test_enc)

# Confusion Matrix
cm = confusion_matrix(y_test_enc, y_pred)

plt.figure(figsize=(10,6))
sns.heatmap(pd.DataFrame(cm),
            annot=True,
            fmt="d",
            xticklabels=le.classes_,
            yticklabels=le.classes_)
plt.title("Confusion Matrix")
plt.ylabel("Gerçek")
plt.xlabel("Tahmin")
plt.show()

print("Accuracy:", metrics.accuracy_score(y_test_enc, y_pred))
```
